```python
import pandas as pd
import requests
import holidays
from pathlib import Path
```

## Importaremos el archivo de datos

```python
archivo = "../data/raw/Billing export from 2023-01-01 to 2025-12-19 for SIAMO LA PIZZA.csv"
df = pd.read_csv(archivo)
df
```

```python
df = df.iloc[:-1]
```

```python
df["activationTime"] = pd.to_datetime(df["activationTime"], format="%d/%m/%Y", errors="coerce")

df = df.sort_values("activationTime")

df
```

```python
df = df.rename(columns={"activationTime": "date"})
```

```python
df
```

```python
df.dtypes
```

```python
cols_to_drop = ["firstBillNumber", "lastBillNumber", "taxableBase", "totalWithDiscount", "discountsTotal", "tip", "cash", "tax 10%", "tax 21%", "card", "delitbee", "gloriafood", "glovo", "justeat", "shop", "the fork qr", "uber"] 
df = df.drop(columns=cols_to_drop)
```

```python
df
```

```python
df["date"] = pd.to_datetime(df["date"])

# 2) Número de día (Lun=1 ... Dom=7)  ✅
df["weekday_num"] = df["date"].dt.isocalendar().day.astype(int)


df.head()
```

```python
postal_code = "28017"
geo_url = "https://geocoding-api.open-meteo.com/v1/search"
geo_params = {
    "name": postal_code,
    "count": 5,
    "language": "es",
    "format": "json"
}
```

```python
g = requests.get(geo_url, params=geo_params, timeout=30)
g.raise_for_status()
results = g.json().get("results", [])
```

```python
results_es = [r for r in results if r.get("country_code") == "ES"] or results
best = results_es[0]  # el primero suele ser el más relevante
```

```python
lat, lon = best["latitude"], best["longitude"]
print("Usando:", best.get("name"), best.get("admin1"), best.get("country"), "->", lat, lon)
```

```python
start = df["date"].min().date().isoformat()
end   = df["date"].max().date().isoformat()

weather_url = "https://archive-api.open-meteo.com/v1/archive"
weather_params = {
    "latitude": lat,
    "longitude": lon,
    "start_date": start,
    "end_date": end,
    "daily": "temperature_2m_max,temperature_2m_min,precipitation_sum,windspeed_10m_max",
    "timezone": "Europe/Madrid"
}

r = requests.get(weather_url, params=weather_params, timeout=30)
r.raise_for_status()
daily = r.json()["daily"]

weather = pd.DataFrame(daily).rename(columns={"time": "date"})
weather["date"] = pd.to_datetime(weather["date"])

df = df.merge(weather, on="date", how="left")

df.head()
```

```python
df["date"] = pd.to_datetime(df["date"])

years = range(int(df["date"].dt.year.min()), int(df["date"].dt.year.max()) + 1)

# Madrid (Comunidad)
es_holidays = holidays.country_holidays("ES", subdiv="MD", years=years)

holiday_dates = set(es_holidays.keys())  # esto son datetime.date

df["is_holiday"] = df["date"].dt.date.isin(holiday_dates)
df["is_holiday_eve"] = (df["date"] + pd.Timedelta(days=1)).dt.date.isin(holiday_dates)

df.head()
```

```python
df["is_holiday_eve"] = (df["date"] + pd.Timedelta(days=1)).dt.date.isin(holiday_dates)
```

```python
df["is_holiday_after"] = (df["date"] - pd.Timedelta(days=1)).dt.date.isin(holiday_dates)
```
